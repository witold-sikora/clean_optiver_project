```python
import numpy as np
import pandas as pd
import torch 
```

```python
from models.residual_nn import MLP
```

```python
model = MLP(in_dim=6, hidden_dim=128, num_blocks=3, dropout=0.05, emb_dim=16, emb_num=127)
model.load_state_dict(torch.load("first_model_loss_0.2451.pth"))
```

```python
df = pd.read_csv('../baseline3.csv', index_col=False).iloc[:, 1:]
df['target'] = df['target'] * 1000
df['sigma'] = df['sigma'] * 1000

X = df.loc[:, ['sigma', 'spread', 'log_volume', 'log_time_diff', 'shift_ratio', 'book_density_abs_impact']]
stock = df.loc[:, ['stock_id']]
y = df['target']

X = X.to_numpy()
stock = stock.to_numpy()
y = y.to_numpy()

X = torch.tensor(X, dtype=torch.float32)
stock = torch.tensor(stock, dtype=torch.int).squeeze(-1)
y = torch.tensor(y, dtype=torch.float32)
```

```python
model.eval()
model = model.cpu()
with torch.no_grad():
    preds = model(X, stock)
    print(MLP.criterion(preds, y))
```
